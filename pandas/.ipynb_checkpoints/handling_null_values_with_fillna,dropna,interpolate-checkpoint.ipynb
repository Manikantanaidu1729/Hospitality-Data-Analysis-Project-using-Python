```python
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df.set_index("day",inplace=True)
df
```

```python
df.fillna(method="ffill")  # ffill = forward fill
```

```python
df.fillna(method="bfill") #bfill = backward fill
```

```python
df.fillna(method="bfill",axis=1) #bfill = backward fill
```

```python
df
```

```python
df.fillna(method="ffill",limit=1)
```

```python
df.fillna(method="ffill",limit=2)
```

```python
df
```

```python
df.interpolate() # it linearly interpolate the null values i.e. linearly increasing
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(how="all")
```

```python
df.dropna(how="any")  # if any one of the column is null then it removes the row
```

```python
df.dropna(thresh=2)  # only keep rows with atleast 2 non-NaN values
```

```python
df
```

```python
df.dropna(thresh=1)    # only keep rows with atleast 1 non-NaN values
```
