```python
def grouper(df,idx,col):
    if 80<=df[col].loc[idx] <=90:
        return "80-90"
    elif 50 <= df[col].loc[idx] <=60:
        return "50-60"
    else:
        return "others"
```

```python
g = df.groupby(lambda idx : grouper(df,idx,"temperature"))
```

```python
df.groupby(lambda idx : grouper(df,idx,"temperature")).describe()
```

```python
g
```

```python
for key,d in g:
    print("key: ",key)
    print()
    print("data: ",d)
    print()
```
